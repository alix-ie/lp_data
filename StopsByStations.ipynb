```python
import pandas as pd
```

```python
from geopy.distance import geodesic
```

```python
stations = pd.read_json('metro_stations.json', encoding='cp1251')
```

```python
stops = pd.read_csv('bus_stops.csv', encoding='cp1251', sep=';')
```

```python
stations
```

```python
stops
```

```python
stops_by_stations = {station: set() for station in set(stations['NameOfStation'])}
```

```python
stops_coords = {name: (lng, lat) for name, lng, lat in zip(stops['Name'], stops['Longitude_WGS84'], stops['Latitude_WGS84'])}
```

```python
for x, y, z in zip(stations['NameOfStation'], stations['Longitude_WGS84'], stations['Latitude_WGS84']):
    station_coords = (y,z)
    
    for stop in stops_coords:
        if geodesic(station_coords, stops_coords[stop]).km <= 0.5:
            stops_by_stations[x].add(stop)
```

```python
max_stops = max((len(stops_by_stations[station]), station) for station in stops_by_stations)
```
