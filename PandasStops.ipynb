```python
import pandas as pd
```

```python
stops = pd.read_csv(
    open('bus_stops.csv', 'r', encoding='cp1251'),
    sep=';'
)
```

```python
stops.head(10)
```

```python
stops_count = stops.groupby(['Street']).size()
```

```python
stops_count.sort_values(ascending=False, inplace=True)
```
